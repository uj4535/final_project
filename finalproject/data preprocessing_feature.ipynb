```python
def textmining_serious(wordlist, column):
    for word in wordlist:
        df.loc[df['feature'].str.contains(word), column]='s'
    return df.loc[column=='s', ['feature', 'column']]
def textmining_able(wordlist, column):
    for word in wordlist:
        df.loc[df['feature'].str.contains(word), column]='a'
    return df.loc[column=='a', ['feature', 'column']]
```
